```python
def decorator(func):
    def wrapper(*args, **kwargs):
        print("执行前的操作")
        result = func(*args, **kwargs)  # 原函数可以接收参数
        print("执行后的操作")
        return result
    return wrapper

@decorator
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")



///////////////////////////////////////////

装饰器的执行过程拆解
装饰器的核心：decorator(func)

decorator 是一个装饰器函数。
它接收一个参数 func，这个参数就是被装饰的函数，也就是 greet()。
换句话说：

当你写 @decorator 的时候，Python 相当于在背后运行了这一行代码：
python
复制代码
greet = decorator(greet)
decorator 返回一个新的函数 wrapper()，它替换了原来的 greet()。
wrapper(*args, **kwargs) 的作用

wrapper() 是装饰器返回的新函数，它包裹了原函数 greet()。
*args 和 **kwargs 是可变参数，用来接收原函数的所有位置参数和关键字参数。
比如调用 greet("Alice") 时，args 是 ("Alice",)，kwargs 是空字典 {}。
所以：

wrapper 是一个“包装层”，可以在调用原函数 greet 的前后，添加额外的功能（比如打印日志）。
func(*args, **kwargs) 的作用

func 是 decorator(func) 中的参数，它就是原函数 greet()。
当你写 func(*args, **kwargs) 时，实际上就是调用了 greet(name)。
```

```python
委托
委托是一种设计模式，它将一个对象的职责或行为委托给另一个对象来执行。
本质上，委托是一种代理行为，即调用者将任务转交给特定的处理者（被委托对象）

class TaskExecutor:
    def run(self, task_name):
        print(f"Executing task: {task_name}")

class Client:
    def __init__(self, executor):
        self.executor = executor  # 委托给 executor 处理

    def perform_task(self, task_name):
        print("Client delegating task...")
        self.executor.run(task_name)

executor = TaskExecutor()
client = Client(executor)
client.perform_task("Data Processing")
```
